```python
import pandas as pd

df = pd.read_csv('final_data.csv')
df
```

```python
#1. Який середній вік футболістів?
average_age = df['age'].mean()

print(f"Середній вік футболістів: {round(average_age)} років")
```

```python
#2. Скільки жовтих карток отримали гравці на позиції воротаря?")
yellow_cards_defenders = df[df['position'] == 'Goalkeeper']['yellow cards'].sum()
print(f"Гравці на позиції воротаря отримали {round(yellow_cards_defenders)} жовтих карток")
```

```python
#3. Показати список найкращих гравців молодше 25 років за кількістю нагород.
top_young_players = df[df['age'] < 25].nlargest(10, 'award')[['name','award','current_value']]
print(top_young_players)
```
